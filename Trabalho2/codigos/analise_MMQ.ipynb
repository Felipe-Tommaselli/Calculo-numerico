```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('dados_temperatura_minima.csv')
```

```python
dados.head()
```

```python
X = dados['Mês'].values
Y = dados['Temperatura Mínima'].values
```

```python
plt.scatter(X,Y,label='T(t)', c="g");
plt.xlabel('t (tempo)');
plt.ylabel('T (Temperatura mínima)');
plt.legend();
```

```python
modelo3 = np.poly1d(np.polyfit(X, Y, 3))
modelo4 = np.poly1d(np.polyfit(X, Y, 4))
modelo5 = np.poly1d(np.polyfit(X, Y, 5))
modelo7 = np.poly1d(np.polyfit(X, Y, 7))
modelo10 = np.poly1d(np.polyfit(X, Y, 10))
modelo100 = np.poly1d(np.polyfit(X, Y, 100))
```

```python
print("grau 3:", np.polyfit(X, Y, 3))
print("grau 4:", np.polyfit(X, Y, 4))
print("grau 5:", np.polyfit(X, Y, 5))
```

```python
y3 = modelo3(X)
y4 = modelo4(X)
y5 = modelo5(X)
y7 = modelo7(X)
y10 = modelo10(X)
y100 = modelo100(X)
```

```python
fig, ax = plt.subplots(2, 3)
ax[0,0].scatter(X,Y, c="g");
ax[0,0].plot(X,y3,color='red');
ax[0,0].set_title('Grau 3');

ax[1,0].scatter(X,Y, c="g");
ax[1,0].plot(X,y4,color='red');
ax[1,0].set_title('Grau 4');

ax[0,1].scatter(X,Y, c="g");
ax[0,1].plot(X,y5,color='red');
ax[0,1].set_title('Grau 5');

ax[1,1].scatter(X,Y, c="g");
ax[1,1].plot(X,y7,color='red');
ax[1,1].set_title('Grau 7');
plt.tight_layout();

ax[0,2].scatter(X,Y, c="g");
ax[0,2].plot(X,y10,color='red');
ax[0,2].set_title('Grau 10');
plt.tight_layout();

ax[1,2].scatter(X,Y, c="g");
ax[1,2].plot(X,y100,color='red');
ax[1,2].set_title('Grau 100');
plt.tight_layout();
```

```python
plt.scatter(X,Y,label='T(t)', c="g");
plt.plot(X,y5,color='red');
plt.xlabel('t (tempo)');
plt.ylabel('T (Temperatura mínima)');
plt.legend();
```

```python
a = (11.64408333, -0.07047697082, 0.2184961566, -0.08537618565, 0.2602394922, -0.04420947947);
b = (0.0000000000, 0.2367753488, 0.2352830559, 0.153381605, -0.2641546302, -0.0710746394);

y_trig2 = a[0] + a[1]*np.sin(1*X*2*np.pi/120
                ) + b[1]*np.cos(1*X*2*np.pi/120
                ) + a[2]*np.sin(2*X*2*np.pi/120
                ) + b[2]*np.cos(2*X*2*np.pi/120);

y_trig3 = a[0] + a[1]*np.sin(1*X*2*np.pi/120
                ) + b[1]*np.cos(1*X*2*np.pi/120
                ) + a[2]*np.sin(2*X*2*np.pi/120
                ) + b[2]*np.cos(2*X*2*np.pi/120
                ) + a[3]*np.sin(3*X*2*np.pi/120
                ) + b[3]*np.cos(3*X*2*np.pi/120);

y_trig4 = a[0] + a[1]*np.sin(1*X*2*np.pi/120
                ) + b[1]*np.cos(1*X*2*np.pi/120
                ) + a[2]*np.sin(2*X*2*np.pi/120
                ) + b[2]*np.cos(2*X*2*np.pi/120
                ) + a[3]*np.sin(3*X*2*np.pi/120
                ) + b[3]*np.cos(3*X*2*np.pi/120
                ) + a[4]*np.sin(4*X*2*np.pi/120
                ) + b[4]*np.cos(4*X*2*np.pi/120);

y_trig5 = a[0] + a[1]*np.sin(1*X*2*np.pi/120
                ) + b[1]*np.cos(1*X*2*np.pi/120
                ) + a[2]*np.sin(2*X*2*np.pi/120
                ) + b[2]*np.cos(2*X*2*np.pi/120
                ) + a[3]*np.sin(3*X*2*np.pi/120
                ) + b[3]*np.cos(3*X*2*np.pi/120
                ) + a[4]*np.sin(4*X*2*np.pi/120
                ) + b[4]*np.cos(4*X*2*np.pi/120
                ) + a[5]*np.sin(5*X*2*np.pi/120
                ) + b[5]*np.cos(5*X*2*np.pi/120);
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
MAE3 = mean_absolute_error(Y,y3)
MAE4 = mean_absolute_error(Y,y4)
MAE5 = mean_absolute_error(Y,y5)
MAE7 = mean_absolute_error(Y,y7)
MAE10 = mean_absolute_error(Y,y10)
MAE100 = mean_absolute_error(Y,y100)


print("MAE (grau=3) = {:0.4f}".format(MAE3))
print("MAE (grau=4) = {:0.4f}".format(MAE4))
print("MAE (grau=5) = {:0.4f}".format(MAE5))
print("MAE (grau=7) = {:0.4f}".format(MAE7))
print("MAE (grau=10) = {:0.4f}".format(MAE10))
print("MAE (grau=100) = {:0.4f}".format(MAE100))
```

```python
RMSE3 = np.sqrt(mean_squared_error(Y,y3))
RMSE4 = np.sqrt(mean_squared_error(Y,y4))
RMSE5 = np.sqrt(mean_squared_error(Y,y5))
RMSE7 = np.sqrt(mean_squared_error(Y,y7))
RMSE10 = np.sqrt(mean_squared_error(Y,y10))
RMSE100 = np.sqrt(mean_squared_error(Y,y100))
RMSE2_trig = np.sqrt(mean_squared_error(Y,y_trig2))
RMSE3_trig = np.sqrt(mean_squared_error(Y,y_trig3))
RMSE4_trig = np.sqrt(mean_squared_error(Y,y_trig4))
RMSE5_trig = np.sqrt(mean_squared_error(Y,y_trig5))

print("RMSE (grau=3) = {:0.4f}".format(RMSE3))
print("RMSE (grau=4) = {:0.4f}".format(RMSE4))
print("RMSE (grau=5) = {:0.4f}".format(RMSE5))
print("RMSE (grau=7) = {:0.4f}".format(RMSE7))
print("RMSE (grau=10) = {:0.4f}".format(RMSE10))
print("RMSE (grau=100) = {:0.4f}".format(RMSE100))
print("RMSE Trig (grau=2) = {:0.4f}".format(RMSE2_trig))
print("RMSE Trig (grau=3) = {:0.4f}".format(RMSE3_trig))
print("RMSE Trig (grau=4) = {:0.4f}".format(RMSE4_trig))
print("RMSE Trig (grau=5) = {:0.4f}".format(RMSE5_trig))
```
